## 1. Sistemas M/M/s

Tomando en consideración los datos del ejemplo del ejercicio, se solicita inicialmente determinar el número de servidores, necesarios para cumplir con todas las especificaciones. El ejemplo dado por su parte, responde a un sistema de tipo **M/M/s** debido a que se solicita encontrar el número de $s$ servidores. En un sistema **M/M/s**, las llegadas son descritas por un proceso de Poisson con parámetro λ. Cada uno de los s servidores tiene un tiempo de servicio exponencial con parámetro ν.
El proceso X(t) = i (notación abreviada Xt) describe el estado del sistema al tiempo t para S = {0, 1, . . .} (es decir, i ≥ 0). Además, se dice que el sistema alcanza un estado estacionario cuando  ρ = λ/(sν) < 1. 


 Para esta variedad de procesos, se tienen las siguientes ecuaciones: 


\begin{gathered}
\Omega_{0}=\lambda \quad \Omega_{i \rightarrow i-1}= \begin{cases}i \nu & \text { para } 0<i<s \\
s \nu & \text { para } i \geq s\end{cases} \\
\phi_{0}=\left[\sum_{k=0}^{s-1} \frac{(s \rho)^{k}}{k !}+\frac{(s \rho)^{s}}{s !(1-\rho)}\right]^{-1} \\
\phi_{i}= \begin{cases}\frac{(s \rho)^{i}}{i !} \phi_{0} & \text { para } i<s \\
\frac{s^{s} \rho^{i}}{s !} \phi_{0} & \text { para } i \geq s\end{cases}
\end{gathered}


Asimismo, es posible describir el número promedio de clientes, como:


$$L=\frac{\lambda}{\nu}+\frac{\rho(s \rho)^{s}}{s !(1-\rho)^{2}} \phi_{0}$$


Y el número promedio de clientes en una fila (cuando todos los servidores están
ocupados), tal que:

$$
L_{q}=\frac{\rho(s \rho)^{s}}{s !(1-\rho)^{2}} \phi_{0}=L-\frac{\lambda}{\nu}
$$


### 1.1. - Problema de ejemplo: número de servidores

> Es menester tomar una decisión con respecto al número de "servidores" necesarios para cierto proceso de vacunación, de forma tal que el sistema no exceda 100 personas en fila durante el 95% del tiempo de servicio. 

>En un estudio de afluencia y del tiempo de servicio se asumen y determinan los siguientes parámetros:

>- La tasa de llegada $\lambda$ es constante durante ocho horas de funcionamiento y tiene un valor de $\lambda = 7$ personas por minuto.

>- La tasa de servicio $\nu$ de cada "servidor" (es decir, personal de salud tomando datos y vacunando) es constante durante ocho horas de funcionamiento y tiene un valor de $\nu = 0.25$ personas por minuto.


Ahora bien, al resolver  el número de servidores se tiene que:


\begin{gathered}

\rho = \frac{\lambda}{sv}  \\

\end{gathered}

Como se sabe que la tasa de llegada de personas por minuto es de $\lambda = 7$ y la tasa de servicio por minuto de cada servidor tiene un valor de $\nu = 0.25$, sustituyendo los valores numéricos en la ecuación anterior:

\begin{gathered}

\rho = \frac{7}{s(0.25)} = \frac{28}{s} \\

\end{gathered}

Ahora bien, se sabe que este valor alcanza un estado estacionario cuando es menor a 1. De ahí, que se debe considerar cuando existe una ruptura en menor medida. Tomando en cuanta la siguiente ecuación: 


$$
P( \text{101 o más clientes en el sistema} ) = \sum_{i=101}^{\infty} (1 - \rho) \rho^i  = 1 - \sum_{i=0}^{100} (1 - \rho) \rho^i = \rho^{101}
$$


De los datos del problema, para tener una fila de 100 o menos personas el 95% (generando un 5% la condición de ruptura; es decir 1-95%) del tiempo se necesita despejar la cantidad total de servidores como sigue:


\begin{gathered}
\rho^{101} \leq 0.05 \\
\left(\frac{28}{s}\right)^{101} \leq 0.05 \\
\left(\frac{28101}{s^{101}}\right) \leq 0.05 \\
\left(\frac{28^{101}}{0.05}\right) \leq s^{101} \\
\left(\frac{28^{101}}{0.05}\right) \leq s^{101} \\
\sqrt[101]{\left(\frac{28101}{0.05}\right)} \leq s \\
28 \cdot \sqrt[101]{\left(\frac{1}{0.05}\right)} \leq s \\
28.842939 \leq s
\end{gathered}


De ahí que deben existir al menos 28.842939 servidores para poder satisfacer los requisitos. Sin embargo, como los servidores son números enteros, se redondea a 29 (su entero inmediato superior).
